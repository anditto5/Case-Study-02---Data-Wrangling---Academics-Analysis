```python
import pandas as pd

mengajar = pd.read_csv("C:/Users/gito2/Downloads/mengajar.csv", encoding='utf8')
dosen = pd.read_csv("C:/Users/gito2/Downloads/tb_dosen.csv", encoding='utf8')
kelas = pd.read_csv("C:/Users/gito2/Downloads/tb_kelas.csv", encoding='utf8')
krs = pd.read_csv("C:/Users/gito2/Downloads/tb_krs.csv", encoding='utf8')
mahasiswa = pd.read_csv("C:/Users/gito2/Downloads/tb_mahasiswa.csv", encoding='utf8')
mata_kuliah = pd.read_csv("C:/Users/gito2/Downloads/tb_mata_kuliah.csv", encoding='utf8')
nilai = pd.read_csv("C:/Users/gito2/Downloads/tb_nilai.csv", encoding='utf8')
print (mengajar.shape, dosen.shape, kelas.shape, krs.shape, mahasiswa.shape)
print (mata_kuliah.shape, nilai.shape, mengajar.shape)
mengajar
```

```python
nilai
```

```python
rerata_nilai = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
rerata_nilai.rename(columns={'nilai': 'rerata_nilai'}, inplace=True)
rerata_nilai
```

```python
var_nilai = nilai.groupby(nilai['NIDN']).var()[['nilai']]
var_nilai.rename(columns={'nilai': 'var_nilai'}, inplace=True)
var_nilai
```

```python
summary_nilai = pd.concat([rerata_nilai, var_nilai], axis=1) # , keys=['level1', 'level2']
summary_nilai
```
